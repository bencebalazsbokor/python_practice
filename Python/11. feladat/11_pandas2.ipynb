4. Hozz létre egy új oszlopot, amely a szöveggé alakított év és hónap értékekből jön létre (pl.: 2021-05). Csoportosítsd az adatokat az újonnan létrejött oszlop és az ország oszlop alapján. Vedd a New_deaths értékek összegét.

```python
import pandas as pd
df = pd.read_csv('source/covid.csv')

df['date'] = pd.to_datetime(df['date'])
df['date_month'] = df['date'].dt.year.astype(str) + '-' + df['date'].dt.month.astype(str).str.zfill(2)
ddf = df.groupby(['country', 'date_month'])['New_deaths'].sum().reset_index()
display(ddf)
```

5. Készíts egy pivot táblát az így létrejött csoportosításból. A sorok az egyes országok legyen az oszlopok pedig az év_hónap oszlop. A pivot tábla többszintű oszlopindexét lapítsd ki a list comprehension módszer segítségével.

```python
import pandas as pd
df = pd.read_csv('source/covid.csv')

df['date'] = pd.to_datetime(df['date'])
df['date_month'] = df['date'].dt.year.astype(str) + '-' + df['date'].dt.month.astype(str).str.zfill(2)
ddf = df.groupby(['country', 'date_month'])['New_deaths'].sum().reset_index()

ddf_pivot = pd.pivot_table(ddf, index=['country'], values=['New_deaths'], columns=['date_month']).fillna(0)
ddf_pivot.reset_index(inplace=True)
ddf_pivot.columns = [col[1].strip() if col[0] == 'New_deaths' else col[0] for col in ddf_pivot.columns.values]
display(ddf_pivot)
```

6. Az így létrejött táblához mergeld hozzá az eredeti tábla iso_code értékeit. Figyelj arra, hogy az eredeti tábla minden országból csak egyet tartalmazzon.

```python
import pandas as pd
df = pd.read_csv('source/covid.csv')

df['iso_code'] = df['iso_code'].astype('category')

df['date'] = pd.to_datetime(df['date'])
df['date_month'] = df['date'].dt.year.astype(str) + '-' + df['date'].dt.month.astype(str).str.zfill(2)
ddf = df.groupby(['country', 'date_month'])['New_deaths'].sum().reset_index()

ddf_pivot = pd.pivot_table(ddf, index=['country'], values=['New_deaths'], columns=['date_month']).fillna(0)
ddf_pivot.reset_index(inplace=True)
ddf_pivot.columns = [col[1].strip() if col[0] == 'New_deaths' else col[0] for col in ddf_pivot.columns.values]

merged = ddf_pivot.merge(df[['iso_code', 'country']].drop_duplicates(), on='country', how='inner')

display(merged)
```
