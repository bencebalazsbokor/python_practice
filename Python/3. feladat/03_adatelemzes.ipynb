```python
FILE_NAME = "szamok.txt"

# Fájlból olvsó függvény
def read_file(path):
    with open(path, "r") as f:
        lines = f.readlines()
    numbers = []
    for l in lines:
        numbers.append(int(l.strip()))
    return numbers
    
# Fájlba író függvény
def write_file(path, content):
    text_content = []
    for c in content:
        text_content.append(str(c))
    output = "\n".join(text_content)
    with open(path, "w") as f:        
        f.write(output)

# Keresés függvény
def search_value():
    value = input("Adja meg a keresett értéket: ")
    
    if(value == ""):
        return
    
    if(not(value.isnumeric())):
        return "Nem egész számot adott meg"
    
    value = int(value)
    index = -1
    for i in range(0, len(numbers)):
        if(numbers[i] == value):
            index = i
            break
    if(index < 0):
        return "A megadott érték nem található!"
    else:
        print(f"A megadott érték indexe: {index}")
        return

# Módosítás függvény index bekérő része
def modify_index():
    index = input("Adja meg a kiválasztott indexet: ")
    
    if(index == ""):
        return
    
    if(not(index.isnumeric())):
        return "Nem egész számot adott meg"
    
    index = int(index)
    if(index >= len(numbers)):
        return "Nem létező index"
    
    while True:
        result = modify_value(index)
        if(result == None):
                break
        else:
            print(result)
    
# Módosítás függvény érték bekérő része
def modify_value(index):
    current_value = numbers[index]  
    message = f"Az aktuális érték: {current_value}\n"
    message += "Milyen értékre szeretné módosítani?\n"
    input_value = input(message)
    
    if(input_value == ""):
        return
    
    if(not(input_value.isnumeric())):
        return "Nem egész számot adott meg"
    
    input_value = int(input_value)
    numbers[index] = input_value
    
    print("Módosítás sikeres!") 
    

# Fő program
numbers = read_file(FILE_NAME)

print("100 SZÁM ALKALMAZÁS")

while True:
    message = "\nVálassza ki a kívánt funkciót:\n"
    message += "\tM - Szám módosítása\n"
    message += "\tK - Szám index keresése\n"
    message += "\tQ - Mentés és kilépés\n"
    
    selected = input(message).upper()
    if selected == "Q":
        write_file(FILE_NAME, numbers)
        print("Módosítások mentve!")
        break
    elif selected == "K":
        while True:
            result = search_value()        
            if(result == None):
                break
            else:
                print(result)
    elif selected == "M":        
        while True:
            result = modify_index()        
            if(result == None):
                break
            else:
                print(result)   
    else:
        print("Nem megfelelő funkció!")
```
